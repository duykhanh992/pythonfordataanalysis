```python
#Get the list
name = input("Enter file:")
if len(name) < 1: name = '53.txt'
handle = open(name)
print(handle)
```

```python
lst = []
for line in handle:
    line = line.rstrip().lower()
    words = line.split()
    for word in words:
        if word not in stopwords:
            lst.append(word)
```
